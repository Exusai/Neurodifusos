```
import numpy as np
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output
from IPython.display import Image
from mpl_toolkits.mplot3d import Axes3D
from numpy import linalg as LA
import matplotlib.image as mpimg
```

```
#Neuron class
class neuron():
  def __init__(self,conn,actf):
    self.conn = conn
    self.actf = actf
    self.bias = np.random.rand(1, 1) * 2 - 1
    self.w = np.random.rand(1, conn) * 2 - 1
    #self.wt = np.transpose(self.w)

  def forwardpass(x,y):
    p = np.array([x,y])
    p.transpose
    z = self.N.w@p+self.N.bias
    a = self.N.actf(z)
    return a
  

#actf
relu  = lambda x: np.maximum(0,x)

#purelin  = lambda x: np.maximum(0,x)

purelin = lambda x: 1 / (1 + np.e ** (-x))

def hardlim(x):
  return 1 if x>0 else 0

def hardlims(x):
  return 1 if x>0 else -1

def Forwardpass(x,y):
    p = np.array([x,y])
    p.transpose
    z = N.w@p+N.bias
    a = N.actf(z)
    return a

def Forwardpass2(x,y):
    p = np.array([x,y])
    p.transpose
    z = N2.w@p+N2.bias
    a = N2.actf(z)
    return a


```

```
ress = 100
X = np.linspace(0,6,ress) 
Y = np.linspace(0,6,ress) 
```

```
px = [1,1,2,2] #class 0
py = [4,5,4,5]

p1x = [3,3,4,4] #class 1
p1y = [1,2,1,2] 



plt.scatter(px,py, marker = 'o',c = 'r')
plt.scatter(p1x,p1y, marker = 'o',c = 'g')

plt.grid(True)
```

```
out = []
p = [[1, 4],
     [1, 5],
     [2,4],
     [2, 5],
     [3,1],
     [3,2],
     [4,1],
     [1,2],
    ]
p = np.transpose(p)

e = []
e2 = []

t = [0,0,0,0,1,1,1,1]
t2 = [-1,-1,-1,-1,1,1,1,1]
N = neuron(2,hardlim)
N2 = neuron(2,purelin)

print(p)
print(t)
```

```
epochs = int(input())

for g in range(epochs):
  for i in range(len(t)):
    #train
    z = N.w@p[:,i]+N.bias
    a = N.actf(z)
    err = t[i]-a
    e.append(float(err))
    N.w = N.w + err*np.transpose(p[:,i])
    N.bias = N.bias + err
    #train2
    z2 = N2.w@p[:,i]+N2.bias
    a2 = N2.actf(z2)
    err2 = t[i]-a2
    e2.append(float(err2))
    R = (1/2)*p@np.transpose(p)
    evalue = LA.eig(R)
    lr = (1/8*(np.amax(evalue[1])))
    #lr = .20
    N2.w = N2.w + 2*lr*err2*np.transpose(p[:,i])
    N2.bias = N2.bias + 2*lr*err2

fig = plt.figure(figsize=(5, 5), dpi= 80, facecolor='w', edgecolor='k')

x = np.linspace(0,4,ress)
m = -(N.bias / N.w[:,1]) / (N.bias / N.w[:,0])
line = m*x+(N.bias/N.w[:,0])

m2 = -(N2.bias / N2.w[:,1]) / (N2.bias / N2.w[:,0])
line2 = m2*x+(N2.bias/N2.w[:,0])      
plt.plot(x,line[0])
plt.plot(x,line2[0])
      
plt.scatter(px,py, marker = 'o',c = 'r')
plt.scatter(p1x,p1y, marker = 'o',c = 'skyblue')
plt.axis("equal")
plt.grid(True)
plt.show()
error = e
error2 = e2
plt.plot(range(len(error)), error)
plt.show()
plt.plot(range(len(error2)), error2)
plt.show()
```
